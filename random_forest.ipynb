```python
import tensorflow_decision_forests as tfdf
import pandas as pd
from sklearn.model_selection import train_test_split
from datetime import datetime
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report, f1_score
import seaborn as sns 
```

```python
data='/content/drive/MyDrive/Colab Notebooks/Second semester/Data/full_dataset.csv'
```

```python
all_ds = pd.read_csv(data)
all_ds = all_ds.sample(frac=1)
```

```python
all_ds.pop('Unnamed: 0')
```

```python
train_df, test_df =  train_test_split(all_ds, test_size=0.2)
```

```python
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_df, label="subcategory")
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_df, label="subcategory")
```

```python
test_y=test_df.pop('subcategory')
test_y
```

```python
model = tfdf.keras.RandomForestModel()
start = datetime.now()
model.fit(train_ds)
end = datetime.now()
```

```python
print(end-start)
```

```python
modelpath='/content/drive/MyDrive/Colab Notebooks/Second semester/Models/RF/Global'
```

```python
model.save(modelpath)
```

```python
tfdf.model_plotter.plot_model_in_colab(model, tree_idx=0, max_depth=3)
```

```python
import matplotlib.pyplot as plt

logs = model.make_inspector().training_logs()

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot([log.num_trees for log in logs], [log.evaluation.accuracy for log in logs])
plt.xlabel("Number of trees")
plt.ylabel("Accuracy (out-of-bag)")

plt.subplot(1, 2, 2)
plt.plot([log.num_trees for log in logs], [log.evaluation.loss for log in logs])
plt.xlabel("Number of trees")
plt.ylabel("Logloss (out-of-bag)")

plt.show()
```

```python
from tensorflow import keras
model = keras.models.load_model(modelpath)
```

```python
predictions_ds=model.predict(test_ds)
```

```python
predictions_df=pd.DataFrame(predictions_ds)
```

```python
predictions_df['subcategory']=predictions_df.idxmax(1)
col='subcategory'
predictions_df=predictions_df.loc[:, predictions_df.columns == col]
```

```python
#confusion matrix
ax= plt.subplot()
cm = confusion_matrix(test_y, predictions_df)
sns.heatmap(cm, annot=True, ax = ax); 
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
```
